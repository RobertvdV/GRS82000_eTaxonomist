```python
from spacy import displacy
import spacy
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from netgraph import Graph
import numpy as np
import pickle
from tqdm import tqdm
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import seaborn as sns
import re
import collections
import glob
import pandas as pd
nlp = spacy.load('en_core_web_trf')
```

```python
kg = [
    ('Brown bear', 'has_part', 'claw'),
]

df = pd.DataFrame(kg, columns=['source', 'edge', 'target'])

G=nx.from_pandas_edgelist(df, "source", "target", 
                          edge_attr=True, create_using=nx.Graph())

fig, ax = plt.subplots(figsize=(12, 6))
#plt.figure(figsize=(8, 3))
edge_labels = dict(zip(list(zip(df.source, df.target)),
                  df['edge'].tolist()))

node_size = {
    'Brown bear': 4,
    'claw': 3,
}

node_color = {
    'Brown bear': 'darkgreen',
    'claw': 'lightgreen',   
}

node_layout = {
    'Brown bear': np.array([0, 0.5]),
    'claw': np.array([.2, .5]),
}

node_labels = {node : node for node in G}

Graph(G, node_layout=node_layout, edge_layout='curved', 
      node_size=node_size, node_color=node_color, node_labels=node_labels,
      edge_labels=edge_labels, node_label_offset=(0.00, -0.05), 
      node_label_rotate=50, arrows=True, edge_width=.7,
      node_label_fontdict=dict(size=20),
      edge_label_fontdict=dict(size=15),
      ax=ax)

plt.savefig('simple_graph_1.png', format='png', dpi=1200, bbox_inches='tight')
```

```python
kg = [
    ('Brown bear', 'has_part', 'claw'),
    ('Brown bear', 'has_part', 'fur'),
    ('fur', 'has_property', 'brown'),
    ('claw', 'has_property', 'sharp'),
]

df = pd.DataFrame(kg, columns=['source', 'edge', 'target'])

G=nx.from_pandas_edgelist(df, "source", "target", 
                          edge_attr=True, create_using=nx.Graph())

fig, ax = plt.subplots(figsize=(12, 6))
#plt.figure(figsize=(8, 3))
edge_labels = dict(zip(list(zip(df.source, df.target)),
                  df['edge'].tolist()))

node_size = {
    'Brown bear': 4,
    'fur': 3,
    'claw': 3,
    'sharp': 2,
    'brown': 2,
}

node_color = {
    'Brown bear': 'darkgreen',
    'fur': 'lightgreen',
    'claw': 'lightgreen',
    'sharp': 'white',
    'brown': 'white',    
}

node_layout = {
    'Brown bear': np.array([0, 0.5]),
    'fur': np.array([.2, .55]),
    'claw': np.array([.2, .45]),
    'sharp': np.array([.4, .45]),
    'brown': np.array([.4, .55]),
}

node_labels = {node : node for node in G}

Graph(G, node_layout=node_layout, edge_layout='curved', 
      node_size=node_size, node_color=node_color, node_labels=node_labels,
      edge_labels=edge_labels, node_label_offset=(0.00, -0.05), 
      node_label_rotate=50, arrows=True, edge_width=.7,
      node_label_fontdict=dict(size=20),
      edge_label_fontdict=dict(size=15),
      ax=ax)

plt.savefig('simple_graph_2.png', format='png', dpi=1200, bbox_inches='tight')
```
