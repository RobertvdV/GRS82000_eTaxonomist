### Get Plant Glossary List

```python
# URL
URL = 'https://en.wikipedia.org/wiki/Glossary_of_plant_morphology'
# Get the page
page = requests.get(URL, timeout=5)
soup = BeautifulSoup(page.content, "lxml", from_encoding="iso-8859-1")   

glossary = collections.defaultdict(list)
# Find all H4 
for chapter in soup.find_all('h4')[0:]:
    # Clean
    chapter_text = chapter.text.rstrip('[edit]')
    # Find all siblings
    for sibling in chapter.find_next_siblings():
        # Find the parent
        for parent in sibling.find_previous_sibling('h4'):
            # Only append if correspond to current chapter
            if parent.text == chapter_text:
                if 'â' in sibling.text:
                    for tag in sibling.find_all('li'):
                        candidates = tag.text.split('â')[0]
                        candidates = candidates.split('/')
                        for candidate in candidates:
                            glossary[chapter_text.lower()].append(candidate.strip().lower())
```

```python
descriptions_id = collections.defaultdict(list)
descriptions = collections.defaultdict(list)

### PLOTTING ###
parts = []
### PLOTTING ###

compounds = [
    'upper', 'lower', 'apex', 
    'dorsal', 'central', 'lateral'
]

def compound_reconstructor(token, doc):
    if token.i == 0:
        trait = token
    elif doc[token.i - 3].dep_ == 'compound':
        trait = doc[token.i - 3: token.i + 1]
    elif doc[token.i - 3].dep_ in compounds:
        trait = doc[token.i - 3: token.i + 1]
    elif doc[token.i - 2].dep_ == 'compound':
        trait = doc[token.i - 2: token.i + 1]
    elif doc[token.i - 1].dep_ == 'compound':
        trait = doc[token.i - 1: token.i + 1]
    elif doc[token.i - 1].text.lower() in compounds:
        trait = doc[token.i - 1: token.i + 1]
    else:
        trait = token   
    return trait.lemma_

def extract_adjectives(token, doc):
    adjs = []
    span = None
    for child in token.children:
        if child.lemma_ in compounds:
            continue
        elif child.lemma_ in ['wide', 'long'] and child.dep_ == 'amod':
            tree = ' '.join([t.lemma_ for t in child.subtree])
            #tree = tree + ' ' + child.lemma_
            adjs.append(tree)
        elif child.tag_ == 'JJ' and child.dep_ == 'amod':
            if 'nummod' in [t.dep_ for t in child.children]:
                for grandchild in child.children:
                    if grandchild.dep_ == 'nummod':
                        span = grandchild.lemma_ + '-' + child.lemma_
                        adjs.append(span)
            else:
                adjs.append(child.lemma_)
            for grandchild in child.subtree:
                if grandchild.dep_ == 'conj'and grandchild.tag_ == 'JJ':
                    if 'advmod' in [t.dep_ for t in grandchild.children]:
                        for ggchild in grandchild.children:
                            if ggchild.dep_ == 'advmod':
                                span = ggchild.lemma_ + ' ' + grandchild.lemma_
                                adjs.append(span)
                    else:
                        adjs.append(grandchild.lemma_)
        elif child.tag_ == 'VBN' and child.dep_ == 'amod':
            adjs.append(child.text)
        elif child.dep_ == 'conj'and child.tag_ == 'JJ':
            adjs.append(child.lemma_)
            for grandchild in child.subtree:
                if grandchild.dep_ == 'conj'and grandchild.tag_ == 'JJ':
                    if 'advmod' in [t.dep_ for t in grandchild.children]:
                        for ggchild in grandchild.children:
                            if ggchild.dep_ == 'advmod':
                                span = ggchild.lemma_ + ' ' + grandchild.lemma_
                                adjs.append(span)
                    else:
                        adjs.append(grandchild.lemma_)
                    #print(grandchild)
    return adjs

for species in tqdm_notebook(list(data.keys())[333:334]):
    for idx, text in enumerate(data[species][15:16]):
        
        text = text.replace('.', ' ')
        text = f'{text}.'
        
        # Reset variables
        subject = None
        root = None
        span = None
        doc = nlp(text.lower().capitalize())
        descriptions = collections.defaultdict(list)
        
        ### NORMAL SUBJECT ###
        subject = next((t for t in doc if t.dep_ == 'nsubj'), None)
        # Check if subject
        if subject:
            # Get the full span
            trait = compound_reconstructor(subject, doc)
            # Check if exists within glossary
            for base_part, subparts in glossary.items():
                if subject.lemma_.lower() in subparts or subject.text.lower() in subparts:
                    # Append basepart & subpart   
                    descriptions[species].append(('species', 'base part', base_part))
                    descriptions[species].append((base_part, 'part', trait))
                    # Find possible parents
                    parent = next(subject.ancestors, None)
                    # Check if parent
                    if parent:
                        # If object of preposition
                        prep = next((t for t in parent.children if t.dep_ == 'prep'), None)
                        if prep:
                            # Find object
                            child = next((t for t in prep.children if t.pos_ == 'NOUN' if t.dep_ == 'pobj'), None)
                            if child:
                                # Append subpart
                                descriptions[species].append((trait, f'{parent} {prep}', child.lemma_))
                                # Extract Adjectives
                                adjs = extract_adjectives(child, doc)
                                for adj in adjs:
                                    descriptions[species].append((child.lemma_, 'is', adj))
                        else:        
                            descriptions[species].append((trait, 'is', parent.text))
                        
        ### ROOT ###
        root = next((t for t in doc if t.dep_ == 'ROOT' if t.pos_ == 'NOUN'), None)
        if root:
            #print(root)
            # Get the full span
            trait = compound_reconstructor(root, doc)
            # Check if exists within glossary
            for base_part, subparts in glossary.items():
                if root.lemma_.lower() in subparts or root.text.lower() in subparts:
                    # Append basepart & subpart   
                    descriptions[species].append(('species', 'base part', base_part))
                    descriptions[species].append((base_part, 'part', trait))
                    ## ADJ of ROOT ##
                    adjs = extract_adjectives(root, doc)
                    #print(root)
                    for adj in adjs:
                        descriptions[species].append((trait, 'is', adj))            
                    ## APPOS ##
                    # conjunction, subordinating or preposition
                    for child in root.children:
                        if child.dep_ == 'appos':
                            # Define relation
                            rel = 'has' if child.pos_ == 'NOUN' else 'is'
                            # The appos contains nothing
                            if len(list(child.subtree)) == 1:
                                descriptions[species].append((trait, rel, child.lemma_))
                            # The appos contains more information
                            else:
                                appos = doc[child.left_edge.i : child.right_edge.i + 1]
                                descriptions[species].append((trait, rel, appos.text))
                                
        ### other ###
        other = next((t for t in doc if t.pos_ == 'NOUN'), None)
        if other:
            #print(root)
            # Get the full span
            trait = compound_reconstructor(other, doc)
            # Check if exists within glossary
            for base_part, subparts in glossary.items():
                if other.lemma_.lower() in subparts or other.text.lower() in subparts:
                    # Append basepart & subpart   
                    descriptions[species].append(('species', 'base part', base_part))
                    descriptions[species].append((base_part, 'part', trait))
                    ## ADJ of ROOT ##
                    adjs = extract_adjectives(other, doc)                                
                    for adj in adjs:
                        descriptions[species].append((trait, 'is', adj))           
    
        descriptions_id[species].append((idx, descriptions[species]))
```

```python
def clean(text):
    return text.strip().lower()


RDF = collections.defaultdict(list)
for species in descriptions_id.keys():
    for idx, triples in descriptions_id[species]:
        if triples:
            for (obj, rel, subj) in triples:
                subjects = subj.split(',')
                obj = clean(obj)
                rel = clean(rel)
                if len(subjects) == 1:
                    RDF[species].append((obj, rel, subj))
                elif len(subjects) > 1:
                    if any("long" in s for s in subjects) and any("wide" in s for s in subjects):
                        # Correct measurements
                        for subj in subjects:
                            subj = clean(subj)
                            RDF[species].append((obj, rel, subj))
                    else:
                        RDF[species].append((obj, rel, clean(subjects[0])))
                        RDF[species].append((clean(subjects[0]), 'be', clean(subjects[1])))
                    
```

```python
RDF_text = collections.defaultdict(list)
for species in RDF.keys():
    for (obj, rel, subj) in RDF[species]:
        RDF_text[species].append(f'{obj} {rel} {subj}.'.capitalize())
```

```python
source   = []
relation = []
target   = []

for bird in RDF.keys():
    for triple in RDF[bird]:
        if not all(triple):
            continue

        source.append(triple[0])
        relation.append(triple[1])
        target.append(triple[2])
    
kg_df = pd.DataFrame({'source':source, 'target':target, 'edge':relation})
```

```python
species = list(data.keys())
baseparts = list(glossary.keys())

nodes = [(source, target) for source, target in zip(kg_df['source'].values, kg_df['target'].values)]
G=nx.from_pandas_edgelist(kg_df, "source", "target", 
                          edge_attr=True, create_using=nx.Graph())
node_labels = {node : node for idx, node in enumerate(G)}
node_size = {}
node_color = {}
node_labels_large = {}

edge_labels = dict(zip(list(zip(kg_df.source, kg_df.target)),
                  kg_df['edge'].tolist()))

for node in node_labels:
    if node == 'species':
        node_size[node] = 2.2
        node_color[node] = 'darkgreen'
        node_labels_large[node] = node
    elif node in baseparts:
        node_size[node] = 2
        node_color[node] = 'green'
        node_labels_large[node] = node
    elif node in parts:
        node_size[node] = 1.5
        node_color[node] = 'lightgreen'
        node_labels_large[node] = ''
    else:
        node_size[node] = 1.
        node_color[node] = 'white'
        node_labels_large[node] = ''

```

```python
plt.figure(figsize=(15, 15))
Graph(nodes, node_layout='dot', edge_layout='curved', arrows=True,
      node_labels=node_labels, node_label_offset=(0.00, -0.035), node_size=node_size, 
      node_label_fontdict=dict(size=8, rotation=30, ha='right'), node_color=node_color,
      edge_labels=edge_labels, edge_width=0.4, edge_label_fontdict=dict(size=7))

#plt.savefig('plot.pdf', format='pdf', dpi=1200, bbox_inches='tight')
```
