```python
import spacy
import pickle
from spacy import displacy
nlp = spacy.load('en_core_web_trf')
from bs4 import BeautifulSoup
import requests
from tqdm.notebook import tqdm as tqdm_notebook
import collections
import re
import pandas as pd
import numpy as np
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# URL
URL = 'https://en.wikipedia.org/wiki/Glossary_of_bird_terms'
# Get the page
page = requests.get(URL, timeout=5)
soup = BeautifulSoup(page.content, "lxml", from_encoding="iso-8859-1")   
# Find embedded glossary
glossaries = soup.find_all('dt', {'class': 'glossary'})
parts = [part.text.lower().strip() for part in glossaries]
# Get additional anchors ("also know as...")
glossaries_other = soup.find_all('span', {'class': 'anchor'})
parts_other = [part['id'].lower().strip() for part in glossaries_other]
# Append and drop duplicates
parts = list(set((parts + parts_other)))
# Replace underscore with space
glossary = [part.replace('_', ' ') for part in parts]
```

```python
data = pickle.load(open('../../data/description/04_TRAIN_0000000-0002000_BIRDS.pkl', 'rb'))
```

```python
descriptions = collections.defaultdict(list)
textual_list = collections.defaultdict(list)


locations = [
    'upper',
    'lower',
]

trait = ''

for bird in tqdm_notebook(list(data.keys())[0:1], desc='bird'):
    
    for string in data[bird][26:27]:
        doc = nlp(string)
        for token in doc:
            if token.lemma_.lower() in glossary or token.text.lower() in glossary:
                
                trait = token.lemma_.lower()
                print(doc[token.left_edge.i - 1].lemma_)
                if doc[token.left_edge.i - 1].lemma_ in locations:
                    print('yes')
                #print(token)

                ### NORMAL SUBJECTS ###
                if token.dep_ == 'nsubj':
                    for child in next(token.ancestors).children:
                        if child.dep_ == 'acomp':
                            # acompanies
                            descriptions[bird].append((bird, 'has', token.lemma_))
                            descriptions[bird].append((token.lemma_, 'be', child.lemma_))

                            textual_list[bird].append(f'The bird has {token.lemma_}, the {token.lemma_} is {child.lemma_}.')

                for child in token.children:
                    ### ADJ of NOUN ###
                    if child.pos_ in ['ADJ', 'VERB']:
                        # Skip
                        if child.dep_ == 'acl':
                            continue
                        # Adjective 
                        descriptions[bird].append((bird, 'has', token.lemma_))
                        descriptions[bird].append((token.lemma_, 'be', child.lemma_)) 

                        textual_list[bird].append(f'The bird has {token.lemma_}, the {token.lemma_} is {child.lemma_}.')


                    ### NUM of NOUN ###
                    elif child.pos_ == 'NUM':
                        # Number
                        descriptions[bird].append((bird, 'has', token.lemma_))
                        descriptions[bird].append((token.lemma_, 'NUM', child.lemma_)) 

                        textual_list[bird].append(f'The bird has {token.lemma_}, the {token.lemma_} number or size {child.lemma_}.')
                        
        textual_list[bird] = list(set(textual_list[bird]))
        
#with open('../../data/processed/PoS_descriptions.pkl', 'wb') as f:
#    pickle.dump(textual_list, f)
```

```python
source   = []
relation = []
target   = []


for triple in descriptions[bird]:
    if not all(triple):
        continue
    source.append(triple[0])
    relation.append(triple[1])
    target.append(triple[2])
    
kg_df = pd.DataFrame({'source':source, 'target':target, 'edge':relation})
```

```python
# Drop empty edges
trait_list = list(kg_df[kg_df.source == 'Common Tern']['target'])
df_test = pd.concat([kg_df[kg_df.source == 'Common Tern'], kg_df[kg_df['source'].isin(trait_list)]])
kg_df = df_test
```

```python
G=nx.from_pandas_edgelist(kg_df, "source", "target", 
                          edge_attr=True, create_using=nx.MultiDiGraph())

plt.figure(figsize=(30, 30))
labels = dict(zip(list(zip(kg_df.source, kg_df.target)),
                  kg_df['edge'].tolist()))
pos = nx.spring_layout(G, k = 1)
nx.draw(G, with_labels=True, node_size=2500, pos=graphviz_layout(G))
nx.draw_networkx_edge_labels(G, pos=graphviz_layout(G), edge_labels=labels,
                                 font_color='red')
plt.show()
#plt.savefig('plot.pdf', format='pdf', dpi=1200, bbox_inches='tight')
```
